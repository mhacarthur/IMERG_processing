```python
import os
import glob
import numpy as np
import clima_anom as ca
import matplotlib.pyplot as plt
import matplotlib.cbook
import matplotlib.gridspec as gridspec
import cartopy
import cartopy.crs as ccrs
```

```python
data_dir = '/mnt/Data/Data/IMERG/raw_month/3IMERG.20201201-S000000-E235959.12.V06B.nc'
data = ca.read_netcdf(data_dir,2)
lat = data['lat']
lon = data['lon']
PRE = data['precipitation'][0,:,:].T
lon2d, lat2d = np.meshgrid(lon, lat)
```

```python
proj = ccrs.PlateCarree()
cmap = plt.cm.Spectral_r
# cmap = plt.cm.YlGnBu

fig = plt.figure(figsize=(4,4),dpi=200)
gs = gridspec.GridSpec(1,1)

ax1 = plt.subplot(gs[0, 0], projection = proj)
ax1.coastlines(linestyle='-',linewidth=0.1,color='black',zorder=4)
ax1.contourf(lon2d, lat2d, PRE, transform = proj, cmap = cmap)
```
