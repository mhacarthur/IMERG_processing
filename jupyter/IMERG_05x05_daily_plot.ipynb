```python
tmp = ca.read_netcdf(list_files[0],0)
lat = tmp['lat']
lon = tmp['lon']

nlat = len(lat)
nlon = len(lon)

lon2d, lat2d = np.meshgrid(lon, lat)
```

```python
monthly_mean = np.zeros([12,nlat,nlon])
monthly_mean[0,:,:] = np.mean(montlhy_mean_01,axis=0)
monthly_mean[1,:,:] = np.mean(montlhy_mean_02,axis=0)
monthly_mean[2,:,:] = np.mean(montlhy_mean_03,axis=0)
monthly_mean[3,:,:] = np.mean(montlhy_mean_04,axis=0)
monthly_mean[4,:,:] = np.mean(montlhy_mean_05,axis=0)
monthly_mean[5,:,:] = np.mean(montlhy_mean_06,axis=0)
monthly_mean[6,:,:] = np.mean(montlhy_mean_07,axis=0)
monthly_mean[7,:,:] = np.mean(montlhy_mean_08,axis=0)
monthly_mean[8,:,:] = np.mean(montlhy_mean_09,axis=0)
monthly_mean[9,:,:] = np.mean(montlhy_mean_10,axis=0)
monthly_mean[10,:,:] = np.mean(montlhy_mean_11,axis=0)
monthly_mean[11,:,:] = np.mean(montlhy_mean_12,axis=0)
```

```python
levels = np.arange(4,64,4)

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'jun', 'Jul', 'Agu', 'Sep', 'Oct', 'Nov', 'Dec']

cmap = plt.cm.Spectral_r
proj = ccrs.PlateCarree()

fig = plt.figure(figsize=(7,9.5),dpi=200)
gs = gridspec.GridSpec(4,3)

count = 0
for row in range(4):
    for col in range(3):

        ax1 = plt.subplot(gs[row, col], projection = proj)
        ax1.coastlines()
        ax1.add_feature(cfeature.COASTLINE,linewidth=0.5,color='black',zorder=4)
        ax1.add_feature(cfeature.BORDERS,linewidth=0.5,color='black',zorder=4)
        ax1.set_extent([-85,-31,-42,8])
        a1 = ax1.contourf(lon2d,lat2d,monthly_mean[count,:,:],cmap=cmap,levels=levels,extend='max')

        gl = ax1.gridlines(crs = proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
        if col == 0:
            gl.left_labels = True
        else:
            gl.left_labels = False

        if row == 3:
            gl.bottom_labels = True
        else:
            gl.bottom_labels = False

        gl.top_labels = False
        gl.right_labels = False
        
        gl.xlabel_style = {'size': 8, 'color': 'black'}
        gl.ylabel_style = {'size': 8, 'color': 'black'}

        if col == 0 and row == 0:
            ax1.set_title('Rainfall montly means', fontsize = 8, loc='left')

        if col == 2 and row == 0:
            ax1.set_title('IMERG 2001-2021', fontsize = 8, loc='right')

        props = dict(boxstyle='round', facecolor='white', alpha=0.5)
        ax1.text(0.03, 0.5, month_names[count], fontsize=8, transform=ax1.transAxes, va='top', bbox=props)

        # plt.colorbar(a1)
        # ax1.set_title(month_names[count], fontsize = 8, loc='left')

        count = count + 1

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, hspace=0.02, wspace=0.02)

cbaxes = fig.add_axes([0.124, 0.06, 0.776, 0.03]) 
cbar = fig.colorbar(a1, cax = cbaxes,orientation="horizontal")
cbar.ax.tick_params(labelsize=8)
cbar.ax.set_title('mm/day',fontsize=7,position=(1., 0))

salida = '../images/montly_mean.png'
print(salida)
plt.savefig(salida,transparent = False,bbox_inches ='tight',pad_inches = 0.01, facecolor='white')
```
